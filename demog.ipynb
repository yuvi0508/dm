```python
demog = pd.read_sas('/content/demog.sas7bdat', encoding = 'latin-1')
```

```python
demog.head()
```

```python
demog['DOMAIN'], demog['STUDYID'], demog['COUNTRY'] = 'DM', 'ABC-PROJECT-001', 'INDIA' # Create DOMAIN, STUDYID and COUNTRY varaibles
demog['SEX'] = demog.sexc.replace(['Male','Female'], ['M','F']) # Create SEX variable
dm1 = demog.rename(columns={'Subject':'SUBJID', 'SiteNumber':'SITEID'}) # Create SUBJID and SITEID variables
dm1['USUBJID'] = dm1.STUDYID + '-' + dm1.SUBJID # Create USUBJID variable
dm1.head()

```

```python
for index, _dm2 in dm1.iterrows():
## Create RACE
  if _dm2['RAIAN_RAW'].__contains__('1'):
     dm1.loc[index, 'RACE'] = 'AMERICAN INDIAN OR ALASKA NATIVE'
  elif _dm2['RINDIA_RAW'].__contains__('1'):
     dm1.loc[index, 'RACE'] = 'ASIAN'
  elif _dm2['RBLACK_RAW'].__contains__('1'):
     dm1.loc[index, 'RACE'] = 'BLACK OR AFRICAN AMERICAN'
  elif _dm2['RWHITE_RAW'].__contains__('1'):
     dm1.loc[index, 'RACE'] = 'WHITE'
  elif _dm2['RNREP_RAW'].__contains__('1'):
     dm1.loc[index, 'RACE'] = 'NOT REPORTED'
  elif _dm2['ROTHER_RAW'].__contains__('1'):
     dm1.loc[index, 'RACE'] = 'OTHER'
  elif _dm2['RUNK_RAW'].__contains__('1'):
     dm1.loc[index, 'RACE'] = 'UNKNOWN'

## Create ETHNIC
  if _dm2['ENHISP_RAW'].__contains__('1'):
     dm1.loc[index, 'ETHNIC'] = 'NOT HISPANIC OR LATINO'
  else:
     dm1.loc[index, 'ETHNIC'] = 'HISPANIC OR LATINO'

dm1
```

```python
ex = pd.read_sas('/content/exp.sas7bdat', encoding = 'latin-1')
ex.head()
```

```python
### Find the first and last exposure date
ex2 = ex.sort_values(by =['Subject','ex_date'], ascending=[True, True] ) # sort by subject and exposure dates
ex3 = ex2.groupby('Subject').first() # pick the first exposure date of the subject
ex4 = ex3.reset_index() # reset index so that subject become a variable.
ex4
```

```python
## Rename the variables
df_ex4 = ex4[['Subject','ex_date']].rename(columns={'Subject':'SUBJID','ex_date':'RFSTDTC'}) # select and rename variables
df_ex4.head()
```

```python
ex5 = ex2.groupby('Subject').last() # pick the last exposure date of the subject
ex5 = ex5.reset_index() # reset index so that subject become a variable.
ex5
```

```python
df_ex5 = ex5[['Subject','ex_date']].rename(columns={'Subject':'SUBJID','ex_date':'RFENDTC'}) # select and rename variables
df_ex5.head()
```

```python
### merge first and last exposure date
df_ex6 = pd.merge(df_ex4, df_ex5, on='SUBJID')
df_ex6
```

```python
### merge exposure data to demog data
dm2 = dm1.sort_values(by =['SUBJID'], ascending=[True] ) # sort by subject and exposure dates
df_dm3 = pd.merge(dm2, df_ex6, on='SUBJID', how='left') # merge exposure information to dm
df_dm3
```

```python
### Prepare demographic data
df_dm4 = df_dm3.drop(columns=['sexc','RINDIA_RAW','RAIAN_RAW','RBLACK_RAW','RWHITE_RAW','RNREP_RAW','ROTHER_RAW','RUNK_RAW','ENHISP_RAW','EOHISP_RAW']) # Drop variables
df_dm4
```

```python
with open('/content/cdisc/dm.xpt', 'wb') as f:
  xport.from_dataframe(df_dm4, f)
```
