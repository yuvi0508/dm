```python
def NAN_rows(df_stat1):
    row = df_stat1.shape[0]
    x = np.empty((1,2,)) # 3 empty row and 2 columns. You can increase according to your original df
    x[:] = np.nan
    df_x = pd.DataFrame( columns = ['PARAM' ,'COL1'])
    for i in range(row):
        temp = np.vstack([df_stat1.iloc[i].tolist(),x])
        df_x = pd.concat([df_x, pd.DataFrame(temp,columns = ['PARAM' ,'COL1'])], axis=0)

    return df_x

df = pd.DataFrame({
   'PARAM' : ['Age'],
    'COL1' : [' ']
})

# print(NAN_rows(df))
df
```

```python
doc = docx.Document()
# Initialise the table
t = doc.add_table(rows=(df_summary.shape[0] + 1), cols=df_summary.shape[1])
# Add the column headings
for j in range(df_summary.shape[1]):
    t.cell(0, j).text = df_summary.columns[j]
# Add the body of the data frame
for i in range(df_summary.shape[0]):
    for j in range(df_summary.shape[1]):
        cell = df_summary.iat[i, j]
        t.cell(i + 1, j).text = str(cell)
# Save the Word doc
doc.save('Demographic Table.docx')
```

```python
# Add in cell borders by setting the table ‘style’ to “TableGrid”:
# Initialise the Word document
doc = docx.Document()
# Initialise the table
doc.add_heading('Table of Demographics', level=1)
t = doc.add_table(rows=1, cols=df_summary.shape[1])
# Add borders
t.style = 'TableGrid'
# Add the column headings
for j in range(df_summary.shape[1]):
    t.cell(0, j).text = df_summary.columns[j]
# Add the body of the data frame
for i in range(df_summary.shape[0]):
    row = t.add_row()
    for j in range(df_summary.shape[1]):
        cell = df_summary.iat[i, j]
        row.cells[j].text = str(cell)
# Save the Word doc
doc.save('Table Demographics.docx')
```
