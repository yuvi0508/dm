```python
# Import required python modules
import pandas as pd
import xport
import numpy as np
```

```python
# Read SDTM DM SAS Transport XPT file
with open('/content/cdisc/dm.xpt', 'rb') as f:
  df_dm = xport.to_columns(f)
  df_dm2 = pd.DataFrame(df_dm)
df_dm2.head()
```

```python
# Create demographic listing
## Select and rename variables
df_dm3 = df_dm2[['SUBJID', 'SEX', 'RACE', 'ETHNIC', 'AGE']].rename(columns={'SUBJID':'Patient ID',
                                                                            'SEX':'Sex', 'RACE':'Race',
                                                                            'ETHNIC':'Ethnicity',
                                                                            'AGE':'Age(year)'})
```

```python
df_dm3
```

```python
# Left alignment of columns
from pandas import DataFrame

def left_align(df_dm3: DataFrame):
    left_aligned_df = df_dm3.style.set_properties(**{'text-align': 'left'})
    left_aligned_df = left_aligned_df.set_table_styles(
        [dict(selector='th', props=[('text-align', 'left')])]
    )
    return left_aligned_df
```

```python
# Set index variable
df_dmlist = df_dm3.set_index('Patient ID', inplace=False)
left_align(df_dmlist)
```

```python
from pandas import DataFrame

def left_align(df_dmlist: DataFrame):
    # Apply styling to a copy of the DataFrame
    left_aligned_df = df_dmlist.copy()
    left_aligned_df = left_aligned_df.style.set_properties(**{'text-align': 'left'})
    left_aligned_df = left_aligned_df.set_table_styles(
        [dict(selector='th', props=[('text-align', 'left')])]
    )
    # Return the styled DataFrame for display and the original DataFrame for manipulation
    return left_aligned_df, df_dmlist

# Get the styled df and the original df
df_dmlist_styled, df_dmlist2 = left_align(df_dmlist)

# Display the styled df
df_dmlist_styled

# Save the original df to a CSV file
df_dmlist2.to_csv('Demographic Listing.csv')
```
